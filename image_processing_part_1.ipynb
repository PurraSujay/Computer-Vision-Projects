**COMPUTER VISION IMAGE PROCESSING BASIC OPERATIONS !**

```python
!pip install opencv-python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

reading and showing the images !

```python
img= cv2.imread('00-puppy.jpg')
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
img_gray = cv2.imread('00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
```

## Resize Images

```python
img_rgb.shape
# width, height, color channels
```

```python
img =cv2.resize(img_rgb,(1300,275))
```

By Ratio 

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img =cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
plt.imshow(new_img)
```

Flipping Images

```python
# Along central x axis
new_img = cv2.flip(new_img,0) #0,1,-1 -------> indicates Fliping left,right,Reverse
plt.imshow(new_img)
```

```python
# Along central y axis
new_img = cv2.flip(new_img,1)
plt.imshow(new_img)
```

```python
# Along both axis
new_img = cv2.flip(new_img,-1)
plt.imshow(new_img)
```

**Drawing on Images**

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

RECTANGLE 

```python
# pt1 = top left
# pt2 = bottom right
cv2.rectangle(blank_img,pt1=(384,0),pt2=(510,128),color=(0,255,0),thickness=5)
```

```python
cv2.rectangle(blank_img,pt1=(100,100),pt2=(400,400),color=(0,0,255),thickness=5)
plt.imshow(blank_img)
```
